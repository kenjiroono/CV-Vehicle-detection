## Background subtraction:
##### Background frame:
To get the background frame, sample frames are randomly taken from the video file and median pixel value for each pixel position is computed from the sample frames. The background frame is converted to gray scale for more efficient processing. 

##### Background subtraction:
The background frame is used to subtract from a given frame so that only the pixels differ in value are retrieved. 
<br> The given frame is converted to gray scale to match the background frame in gray scale then <b>absolute difference(background subtraction)</b> is performed.
<br> The subtracted frame is then binarized and dilated so that contour detection algorithm can perform more efficiently. 


```python
def get_background(cap, sample_size=80):
    """ 
    randomly sample frames of video file and create median gray background image
    
    return: a median gray image
    """
    
    frame_indices = cap.get(cv.CAP_PROP_FRAME_COUNT) * np.random.uniform(size=sample_size)
    frames = []
    
    for idx in frame_indices:
        cap.set(cv.CAP_PROP_POS_FRAMES, idx)
        ret, frame = cap.read()
        frames.append(frame)
        
    # find median pixel values from the sample frames
    median_frame = np.median(frames, axis=0).astype(np.uint8)
    
    return cv.cvtColor(median_frame, cv.COLOR_BGR2GRAY)
```

```python
def background_subtraction(frame, background_frame):
    """
    Perform Background Subtraction by taking absolute difference of given frame and 
    background frame. 
    Binarize and dilate frame for more efficient processing
    
    return: dilated frame
    """
    
    frame_gray = cv.cvtColor(frame, cv.COLOR_BGR2GRAY) # convert frame to gray scale
    frame_diff = cv.absdiff(frame_gray, background_frame) # background subtraction 
    _, frame_thres = cv.threshold(frame_diff, 20, 255, cv.THRESH_BINARY) # binarization
    frame_dilate = cv.dilate(frame_thres, None, iterations=1) # dilate
    
    return frame_dilate
```

```python
# specify video file path
video_file_path = 'Exercise1_Files/Traffic_Laramie_1.mp4'

# load video file
cap = cv.VideoCapture(video_file_path)

# get background
bg_frame = get_background(cap, 50)
```

## Bounding box:
The detected contours are used to get and draw <b>bounding box</b> around the detected area if the contour area is bigger than a threshold. 
<br> The minimum contour area is required to only target vehicle detection which detected area should be bigger than a pedastrian or cyclist. 
<br> To only detect vehicles on a main street, a horizontal detection threshold is set, which is to ignore the vehicles above the main street.

```python
def get_bbox(contours, area_min=3500, area_max=35000, horizontal_limit=250):
    """
    Get bounding box and its coordinated from the contours. 
    Only compute bounding box where contour area is within min~max values.
    Only return bounding box above the horizontal limit that is on the "main street"
    
    return: center points, bounding box coordinates list
    """
    
    detections = [] # store bounding box coordinates and center points
    for contour in contours:
        # contour area limitation 
        if cv.contourArea(contour) < area_min or cv.contourArea(contour) > area_max:
            continue
            
        (x, y, w, h) = cv.boundingRect(contour) # bbox coordinates
        
        # only track vehicle on the "main street"
        if y > horizontal_limit:
            cx, cy = int(x+w/2), int(y+h/2) # center point
            detections.append(((cx, cy), ((x, y, w, h)))) 

    return detections
```

```python
def draw_bbox(frame, detections):
    """ drawing bounding box list to a given frame """
    
    for p in detections:
        x, y, w, h = p[1][0], p[1][1], p[1][2], p[1][3]
        cv.rectangle(frame, (x, y), (x+w, y+h), (50, 255, 0), 2)
        cv.rectangle(frame, (x, y), (x+45, y-30), (50, 255, 0), -1)
```

## Track management:

Object detection works well to detect vehicles, however, it does not consider each object as an unique entity. For this, track list is used to track the detected objects over time with an unique track-id associated with each detected objects. 

The algorithm compares objects in track list and detection.
<br> If the detected object does not exist in the track list, it is added to the track list. 
<br> If the detected object location and object location in the track list are  closer than the maximum distance threshold, the both objects are considered identical and updates the object location in the track list with the detected object location.
<br> If the object in the track list does not find the matching object in the detection, it is removed from the track list.

The track id is drawn above the bounding box.

```python
def object_track(track_list, detections, track_id, max_dist=30):
    """
    Manage object tracking of detected objects.
    Compare detected objects with already detected objects and update its position
    if the distance is less than threshold(min_dist). 
    If the detected object is not associated with any existing object, register as 
    a new exisiting object and assign unique track id.
    
    return: updated track list, detection list, track-id
    """

    # Initialize track list when empty
    if len(track_list) == 0:
        for pt in detections:
            track_list[track_id] = pt
            track_id += 1
    else:
        detections_copy = detections.copy()       
        track_list_copy = track_list.copy()   

        for obj_id, pt2 in track_list_copy.items():
            obj_exists = False
            min_d = max_dist
            min_pt = 0

            for pt in detections_copy:

                # comupte distances between new and exisiting center points
                dist_pts = math.hypot(pt2[0][0] - pt[0][0], pt2[0][1] - pt[0][1])

                # find nearest point between existing and detected point
                if dist_pts < min_d:
                    min_d = dist_pts
                    min_pt = pt
                    obj_exists = True
                    
                    # update detection list
                    if pt in detections:
                        detections.remove(pt)
                        continue         

            # Update track list with new point with least distance
            track_list[obj_id] = min_pt

            # Delete unassociated points
            if not obj_exists:
                track_list.pop(obj_id)

        # Register new points into track list
        for pt in detections:
            track_list[track_id] = pt
            track_id += 1
    
    return track_list, detections, track_id
```

```python
def draw_track_id(frame, track_list):
    """ Iterate track list and draw corresponding track id and center point"""
    
    for obj_id, pt in track_list.items():
        cv.circle(frame, pt[0], 8, (50, 255, 0), -1)
        cv.putText(frame, str(obj_id), (pt[1][0]+6, pt[1][1]-6), 0, 0.9, (255, 255, 255), 3)    
```

## Vehicle counts:
To count the number of vehicles heading to the city center, a rectangle vehicle detection region is defined on  a road. 
<br> If the obejct center points in the track list are inside of the vehicle detection, the vehicle count is incremented. 
<br> To avoid counting for the same vehicle twice as it is possible that the vehicle's center point might still be within the detection boundaries at next frame, once it is detected, the detection is paused for specific frames then resumed, presuming that the vehcile would move pass the detection region during the wait period.   

```python
def count_vehicles(frame, cord, track_list, vehicle_count, vehicle_passed, wait_count, wait_frames=10):
    """
    set up a rectangle to detect center point of vehicle passing through. 
    If detected, wait for wait_frames value and re-activate detection.
    
    return: vehicle_count, wait_count, vehicle_passed(bool)
    """
    
    x,y,w,h = cord[0], cord[1], cord[2], cord[3]
    
    if vehicle_passed:
        cv.rectangle(frame, (x, y), (x+w, y+h), (0, 0, 255), 4)
        cv.putText(frame, ('vehicles = '+str(vehicle_count)), (x-100, y - 10), 0, 1, (0, 0, 255), 3)

    else:
        cv.rectangle(frame, (x, y), (x+w, y+h), (255, 255, 255), 2)
        cv.putText(frame, ('vehicles = '+str(vehicle_count)), (x-100, y-10), 0, 1, (255, 255, 255), 3)            

    # if active, check center point of tracked vehicles are inside of the detection rectangle or not
    if not vehicle_passed:
        for obj_id, pt in track_list.items():
            cv.circle(frame, pt[0], 8, (50, 255, 0), -1)
            cv.putText(frame, str(obj_id), (pt[1][0]+6, pt[1][1]-6), 0, 0.9, (255, 255, 255), 3)    
            
            # center point of vehicle
            cx, cy = pt[0][0], pt[0][1]

            # check either in detection rectangle or not
            if cx > x and cx < x+w and cy < y+h and cy > y:
                vehicle_count += 1
                vehicle_passed = True

    # if wait is over, re-activate the detection
    elif wait_count >= wait_frames:
        vehicle_passed = False
        wait_count = 0

    else:
        wait_count += 1
        
        
    return vehicle_count, wait_count, vehicle_passed        
```

```python
def vehicle_count_summary(vehicle_count, frame_count, fps):
    """
    Calculate total vehicle counts and vehicles per minute
    return: total number of vehicles, cars per minute
    """
    minute = frame_count / fps / 60
    veh_per_min = round(vehicle_count / minute, 2)
    
    print('Total vehicle counts = ', vehicle_count)
    print('vehicles per minute = ', veh_per_min)
```

## Frame differencing:
A specific number of frames processed by background subtraction are added togehter to create a <b>frame differencing</b> frame, which is then used to detect the contours.

```python
frame_diff_nums = 2 # specify number of frame to be used for frame differencing
frame_count = 0 # store frame count
cap = cv.VideoCapture(video_file_path) # reload video file
fps = cap.get(cv.CAP_PROP_FPS) # get the frame per second(fps) of the video file


# _____object detection_____
contours = []
detections = []
area_min = 3500 # minimum contour area to detect  
area_max = 35000 # maximum contour area to detect  
horizontal_limit = 250 # horizontal detection threshold ('detect only main street')


# _____vehicle tracking_____ 
track_list = {}
max_dist = 30 # maximum distance to associate new and existing objects
track_id = 0 # increment each time object is added to the track list


# _____Vehicle counter_____
coord = (540, 350, 30, 120) # coordinates of vehicle pass detector 
vehicle_count = 0 # Increment when vehicle passed
vehicle_passed = False 
wait_frames = 15 # frames to wait while vehicle_passed = True
wait_count = 0 # Increment when vehicle_passed = True
```

```python
while cap.isOpened():

    ret, frame = cap.read()
    
    if ret == True:

        frame_count += 1 
        
        # copy original frame
        frame_copy = frame.copy() 
        
        # Store frames for frame differencing
        if frame_count % frame_diff_nums == 0 or frame_count == 1:
            frame_diff_list = []
        
        # add processed frame to frame_diff_list 
        frame_diff_list.append(background_subtraction(frame, bg_frame))

        # sum up subseaquent frames  
        if len(frame_diff_list) == frame_diff_nums:
            sum_frames = sum(frame_diff_list)
            
            # detect contours around the sum of dilated frame 
            cnts, hierarchy = cv.findContours(sum_frames, cv.RETR_TREE, cv.CHAIN_APPROX_SIMPLE)
            contours = cnts
           
        # Get bounding box from the contours and draw
        detections = get_bbox(contours, area_min, area_max, horizontal_limit)
        
        # draw bounding box
        draw_bbox(frame_copy, detections)
        
        # draw circle at the center point and track-id              
        draw_track_id(frame_copy, track_list)
        
        # object track management
        vt, det, t_id = object_track(track_list, detections, track_id, max_dist)
        track_list = vt
        detections = det
        track_id = t_id
                                 
        # count vehicles heading towards city center
        v_c, w_c, v_p = count_vehicles(frame_copy, coord, track_list, vehicle_count, 
                                       vehicle_passed, wait_count, wait_frames)
        vehicle_count = v_c
        wait_count = w_c
        vehicle_passed = v_p
   

        cv.imshow('vehicle detection & tracking', frame_copy)  

        k = cv.waitKey(1)
        
        # 'escape' key
        if k == 27: 
            break                
        
        # 'space' key
        elif k == 32:
            continue
    
    else:
        break

    
cap.release();
cv.destroyAllWindows()
```

```python
vehicle_count_summary(vehicle_count, frame_count, fps)
```
